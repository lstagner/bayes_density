```python
%pylab inline
import pidly
idl=pidly.IDL('/usr/bin/gdl',idl_prompt='GDL> ')
```

```python
def get_fit_params(fname):
    prefix="/home/lstagner/Dropbox/TESTCASES/"
    idl.filepath=prefix+fname
    idl('restore,filepath')
    idl('params=ne_str.coeff & rhomax=ne_str.rhomax')
    params=idl.params
    rhomax=idl.rhomax
    return params,rhomax

def get_sampling_data(fname):
    prefix="/home/lstagner/Dropbox/TESTCASES/"
    idl.filepath=prefix+fname
    idl('restore,filepath')
    idl('samples=ne_str.sampling_data.samples & log_prob=ne_str.sampling_data.log_prob & logz=ne_str.sampling_data.logz')
    samples=idl.samples
    log_prob=idl.log_prob
    logz=idl.logz
    return samples,log_prob,logz

def ne_function(x,params,rhomax):
    from scipy.special import jn, jn_zeros
    tmp=0.0
    zeros=jn_zeros(0,len(params))
    for ind in range(len(params)):
        tmp = tmp + params[ind]*jn(0,zeros[ind]*x/rhomax)
    return tmp

def plot_bfit(fname):
    params,rhomax =get_fit_params(fname)
    samples,log_prob,logz = get_sampling_data(fname)
    
    prob=exp(log_prob)
    
    rho=linspace(0,rhomax,num=120)
    ne=ne_function(rho,params,rhomax)
    params_most=samples[-1,0:len(params)]
    rhomax_most=samples[-1,len(params):len(params)+1]
    ne_most=ne_function(rho,params_most,rhomax_most[0])
    
    fig,ax=subplots(figsize=(12,8))
    nemax=array([ne.max(),ne_most.max()])
    ax.set_ylim([0,1.1*nemax.max()])
    
    mask = (log_prob > log_prob.max()-6)
    w=where(mask)
    top_probs=prob[w]/prob[w].max()
    ##Plot Density
    for ind in range(len(top_probs)):
        coeff=samples[-ind-1,0:len(params)]
        rhomaxx=samples[-ind-1,len(params):len(params)+1]
        dne=ne_function(rho,coeff,rhomaxx[0])
        ax.plot(rho,dne,color="#BFBFBF", alpha=top_probs[-ind-1])    
    
    ax.plot(rho,ne,'b--',label='Mean',lw=2.0)
    ax.plot(rho,ne_most,'r-.',label='Max',lw=2.0)
    logzstr=r'$log(\mathcal{L}) = %.2f$' % logz
    ax.text(.9,.9*nemax.max(),logzstr,fontsize=20)
    ax.set_xlabel(r'$\rho$', fontsize=18)
    ax.set_ylabel(r'$n_e (10^{19} m^{-3}) $', fontsize=18)
    #ax.set_title(fname,fontsize=20)
    ax.legend(loc=3)
    ax.grid(True)
    fig.savefig("/home/lstagner/latex/bayes_density/figures/"+fname+".eps")
    #Plot Parameter Histograms
    fig2,axes=subplots(nrows=1,ncols=len(samples[0,:]),figsize=(20,3))
    for ind in range(len(axes)):
        subplot(1,len(axes),ind+1)
        s=samples[w,ind].T
        x=s.flatten()
        H2 = hist(x, bins=50,alpha=.3,normed=1)
    fig2.tight_layout
```

```python
plot_bfit("bfit146102.00505_thom4")
plot_bfit("bfit146102.00505_inter2")
plot_bfit("bfit146102.00505_refl3")
plot_bfit("bfit146102.00505_all5")
```
